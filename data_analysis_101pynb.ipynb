```
avg_peak_rpm = df_cars['peak-rpm'].astype('float').mean(axis=0)
print('Average of peak_rpm::=>', avg_peak_rpm)
```

```
df_cars[['bore', 'stroke']] = df_cars[['bore', 'stroke']].astype('float')
df_cars[['normalized-losses']] = df_cars[['normalized-losses']].astype('int')
df_cars[['peak-rpm']] = df_cars[['peak-rpm']].astype('float')
```

```
df_cars.dtypes
```

```
df_cars.head()
```

```
df_cars['city-L/100km'] = 235/df_cars['city-mpg']
```

```
df_cars.head()
```

```
#Data Normalization ::::-
```

```
df_cars['length'] = df_cars['length'] / df_cars['length'].max()
df_cars['width'] = df_cars['width'] / df_cars['width'].max()
```

```
df_cars['horsepower'] = df_cars['horsepower'].astype('int', copy=True)
```

```
%matplotlib inline 
import matplotlib as plt
from matplotlib import pyplot
```

```
plt.pyplot.hist(df_cars['horsepower'])

plt.pyplot.xlabel('horsepower')
plt.pyplot.ylabel('count')
plt.pyplot.title('horsepower bins')
```

```
bins = np.linspace(min(df_cars['horsepower']), max(df_cars['horsepower']), 4)
bins
```

```
group_names = ['low', 'Medium', 'High']
```

```
df_cars['horsepower-binned'] = pd.cut(df_cars['horsepower'], bins, labels=group_names, include_lowest=True )
df_cars[['horsepower','horsepower-binned']].head(20)
```

```
df_cars['horsepower-binned'].value_counts()
```

```
%matplotlib inline
import matplotlib as plt
from matplotlib import pyplot
pyplot.bar(group_names, df_cars["horsepower-binned"].value_counts())

# set x/y labels and plot title
plt.pyplot.xlabel("horsepower")
plt.pyplot.ylabel("count")
plt.pyplot.title("horsepower bins")
```
